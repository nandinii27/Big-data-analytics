**2. Addressing Missing Values**

```python
#CLIENT 


#Cleaning Client Table
#There are no missing values
print("Number of Missing values in clients table: ",client.isna().sum().sum())
# Extracting Birth year
client['birth_year'] = '19' + client['birth_number'].astype(str).str[:2]
client['birth_year'] = client['birth_year'].astype(int)

# Extracting the day of birth
client['birth_day'] = client['birth_number'].astype(str).str[-2:].astype(int)

# Extracting birth month
client['birth_month'] = client['birth_number'].astype(str).str[2:4].astype(int)

# Assigning Gender gender
client['gender'] = 'M'
client.loc[client['birth_month'] > 50, 'gender'] = 'F'

# Recalulating birth month based on gender
client.loc[client['birth_month'] > 50, 'birth_month'] = client.loc[client['birth_month'] > 50, 'birth_month'] - 50

# Calculating Age
client['age'] = 1999 - client['birth_year']

# Assigning Age group
client['age_group'] = client['age'] // 10 * 10


#dropping birth_number column
client=client.drop('birth_number',axis=1)

client.head()
```

```python
#TRANSACTIONS


#number of missing values in transactions table.
print( "Missing values in Transaction table: ",trans.isna().sum().sum())

#Cleaning data in Trans Table
#There are 2278837 missing values

#Type column has info about credit/debit trans
#Date column is the date of transaction
#Operation column is the mode of transaction
#k_symbol is the characterization of the transaction
#account is the account of the partner
#There are 2278837 missing values

#Extracting Account creating date
# 1. Converting date to str
trans['date'] = trans['date'].astype(str)

#2. Converting to date format
trans['transaction_date'] = pd.to_datetime(trans['date'], infer_datetime_format=True)

#Mapping functions to table ( More info in the "Functions" notebook)
trans['trans_type'] = trans['type'].map(typeToEng)
trans['trans_operation'] = trans['operation'].map(operationToEng)
trans['Ksymbol'] = trans['k_symbol'].map(k_symbolToEng)

#dropping the old columns
trans = trans.drop(['type', 'operation','k_symbol','date'], axis=1)

#renaming columns
trans=trans.rename(columns={'account': 'partner_account'})

#Converting partner_account column from float to int
#trans['partner_account'] = trans['partner_account'].astype(int)
trans.head()
```

```python
#ACCOUNT 

#Cleaning Account table
#There are no missing values
print( "Missing values in Account table: ",account.isna().sum().sum())

#frequency column stands for frequency of issuance of statements
#date column is the date of creating of the account


#Mapping function to frequency column
account['statement_frequency'] = account['frequency'].map(FrequencyToEng)

#Extracting Account creating date
# 1. Converting date to str
account['date'] = account['date'].astype(str)

#2. Converting to date format
account['account_creation_date'] = pd.to_datetime(account['date'], infer_datetime_format=True)

#3. Creating Year and Month columns for ease of calculation  -------NEW
account['account_creation_Year'] = account['account_creation_date'].dt.year
account['account_creation_Month'] = account['account_creation_date'].dt.month

#Dropping the old colums
account = account.drop(['date','frequency'], axis=1) 

account.head()
```

```python
#Cleaning Card Table
#There are no missing values
print( "Missing values in card table: ",card.isna().sum().sum())

#type column is the type of card. possible values are "junior", "classic", "gold"
#issues is the issued date in the form YYMMDD

#Extracting issued date
# 1. Converting date to str
card['issued'] = card['issued'].astype(str)

#2. Converting to date format
card['card_issued_date'] = pd.to_datetime(card['issued'], infer_datetime_format=True)

#3. Creating Year and Month columns for ease of calculation -------NEW
card['card_issued_Year'] = pd.DatetimeIndex(card['card_issued_date']).year
card['card_issued_Month'] = pd.DatetimeIndex(card['card_issued_date']).month

#Dropping the old colums
card=card.drop('issued', axis=1)

card.head()
```

```python
#DEMOGRAPHICS

#Missing values
print( "Missing values in district table: ",district.isna().sum().sum())

#Cleaning district data
#There are no missing values

#A1 = District Code
#A2 = District Name
#A3 = Region
#A4 = Numb of inhabitants
#A5 = Numb of Municipalities with inhabitants <499
#A6 = Numb of Municipalities with inhabitants 500-1999
#A7 = Numb of Municipalities with inhabitants 2000-9999
#A8 = Numb of Municipalities with inhabitants >10000
#A9 = Numb of cities
#A10 = Ratio of urban inhabitants
#A11 = Average Salary
#A12 = Unemployment rate 1995
#A13 = Unemployment rate 1996
#A14 = Numb of entrepreneurs per 1000 inhabitants
#A15 = Numb of commited crimes 1995
#A16 = Numb of commited crimes 1996

#Renaming columns
district = district.rename(columns={'A1': 'district_code',
                    'A2':'district_name',
                    'A3':'region',
                    'A4':'inhabitants',
                    'A5': 'municipalities_pop_LT_499',
                    'A6': 'municipalities_pop_LT_1999',
                    'A7': 'municipalities_pop_LT_9999',
                    'A8': 'municipalities_pop_LT_10000',
                    'A9':'numb_cities',
                    'A10': 'ratio_urban_inhab',
                    'A11':'avg_salary', 
                    'A12':'umeployment_rate_95', #has '?' value
                    'A13':'umeployment_rate_96',
                    'A14':'entpr_per_1k_inhab',
                    'A15':'numb_crimes_95', #has an outlier 85677 has '?' value
                    'A16':'numb_crimes_96' #has an outlier 99107
                    })

#Replacing the '?' values in umeployment_rate_95, numb_crimes_95

#1. creating Flag for both the columns
district['umeployment_rate_95_Flag'] = np.where(district['umeployment_rate_95'] == '?', True, False)
district['numb_crimes_95_Flag'] = np.where(district['numb_crimes_95'] == '?', True, False)

#2. Replacing the '?' with NaN and converting type to float
district['umeployment_rate_95'] = district['umeployment_rate_95'].replace('?',np.nan).astype(float)
district['numb_crimes_95'] = district['numb_crimes_95'].replace('?',np.nan).astype(float)

#3. Check the distribution of values in the columns
sns.displot(district['umeployment_rate_95'])
sns.displot(district['numb_crimes_96'])

#4. Basis the distribution, replacing NaN with respective medians of umeployment_rate_95 and numb_crimes_95
district['umeployment_rate_95'] = district['umeployment_rate_95'].replace(np.nan,np.nanmedian(district['umeployment_rate_95']))
district['numb_crimes_95'] = district['numb_crimes_95'].replace(np.nan,np.nanmedian(district['numb_crimes_95']))

district.head()
```

```python
#DISPOSITION

#Cleaning disposition table
#There are no missing values
print( "Missing values in disposition table: ",disp.isna().sum().sum())
#There is no wrong data type or modifications to be made.
disp.head()
```

```python
#ORDER

#Cleaning order table

#There were 1370 NaN values after running the KsymbToEng function initially. 
print( "Missing values in order table: ",order.isna().sum().sum())
#I have substituted np.NaN with No order info in the final else statement.
#As a result there are no missing values

#Has only debit data
#account_id is the account, the order is issued for
# K_symbol is characterization of the payment:
# "POJISTNE" stands for insurance payment 
# "SIPO" stands for household payment
# "LEASING" stands for leasing
# "UVER" stands for loan payment

    
#Mapping the function to the table, we used the same function as in transaction, as k_symbol already existed in that table.
order['KSymbol'] = order['k_symbol'].map(k_symbolToEng)

#Dropping the old column
order = order.drop('k_symbol', axis=1)

order.head()
```

```python
#LOAN

#Cleaning loan table
#There are no missing values
print( "Missing values in loan table: ",loan.isna().sum().sum())


#date is the date when loan was granted
#payments column is the monthly payments
#status is status of paying off the loan
#'A' stands for contract finished, no problems,
#'B' stands for contract finished, loan not payed,
#'C' stands for running contract, OK so far,
#'D' stands for running contract, client in debt

#Extracting the loan grant date
# 1. Converting to string
loan['date'] = loan['date'].astype(str)

#2. Converting to date type
loan['loan_grant_date'] = pd.to_datetime(loan['date'].str[:6], format='%y%m%d')

#3. Creating Year and Month columns for ease of calculation
loan['loan_grant_Year'] = pd.DatetimeIndex(loan['loan_grant_date']).year
loan['loan_grant_Month'] = pd.DatetimeIndex(loan['loan_grant_date']).month

#Mapping the function to column
loan['loan_status'] = loan['status'].map(status)


#dropping the old columns
loan = loan.drop(['date','status'], axis=1)

#renaming columns
loan=loan.rename(columns={'payments': 'monthly_payments',
                     'amount': 'loan_amount',
                     'duration': 'loan_duration'})

loan.head()
```
