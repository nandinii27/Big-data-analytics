```python
import pandas as pd
import numpy as np
```

**Built In Functions**

```python
#Retrieves the fata from a relative path
def ReadData(Dataset, separation=';',lowMemory=True): 
    '''Read file into csv with a sep of ";", proceeds to print related info for the file and returns the created dataframe'''
    df= pd.read_csv(Dataset, sep=separation,low_memory=lowMemory)
    print(df.info())
    return df

#Function to convert Type
def typeToEng(x):
    '''Replace strings in another language into english, in this case for "Type" Column'''
    if x == 'PRIJEM':
        return 'Credit'
    elif x == 'VYDAJ':
        return 'Withdrawal'
    else:
        return np.NaN

#Function to convert Operation
def operationToEng(x):
    '''Replace strings in another language into english, in this case for "Operation" Column'''
    if x == 'VYBER KARTOU':
        return 'Credit Card Withdrawal'
    elif x == 'VKLAD':
        return 'Credit in Cash'
    elif x == 'PREVOD Z UCTU':
        return 'Collection from Another Bank'
    elif x == 'VYBER':
        return 'Withdrawal in Cash'
    elif x == 'PREVOD NA UCET':
        return 'Remittance to Another Bank'    
    else:
        return np.NaN

#Function to convert KSymbol
def k_symbolToEng(x):
    '''Replace strings in another language into english, in this case for "K_Symbol" Column'''
    if x == 'POJISTNE':
        return 'Insurance Payment'
    elif x == 'SLUZBY':
        return 'Payment for Statement'
    elif x == 'UROK':
        return 'Interest Credited'
    elif x == 'SANKC. UROK':
        return 'Sanction Interest if Negative Balance'
    elif x == 'SIPO':
        return 'Household Payment'
    elif x == 'DUCHOD':
        return 'Old Age Pension'  
    elif x == 'UVER':
        return 'Loan Payment' 
    elif x == 'LEASING':
        return 'Leasing'
    else:
        return 'No info' 
#Function to convert frequency to English
def FrequencyToEng(x):
    '''Replace strings in another language into english, in this case for "Frequency" Column'''
    if x == 'POPLATEK MESICNE':
        return 'Monthly Issuance'
    elif x == 'POPLATEK TYDNE':
        return 'Weekly Issuance'
    elif x == 'POPLATEK PO OBRATU':
        return 'Issuance after Transaction'
    else:
        return np.NaN
    

#Function to convert status column
def status(x):
    if x == 'A':
        return 'Contract Finished'
    elif x == 'B':
        return 'Loan not paid - Contract Finished'
    elif x == 'C':
        return 'Running Contract'
    elif x == 'D':
        return 'Client in debt - Running Contract'
    else:
        return pd.Nan
    
    
def preprocess_features(X):
    ''' Preprocesses the football data and converts catagorical variables into dummy variables. '''
    
    # Initialize new output DataFrame
    output = pd.DataFrame(index = X.index)

    # Investigate each feature column for the data
    for col, col_data in X.iteritems():

        # If data type is categorical, convert to dummy variables
        if col_data.dtype == object:
            col_data = pd.get_dummies(col_data, prefix = col)
                    
        # Collect the revised columns
        output = output.join(col_data)
    
    return output
```
